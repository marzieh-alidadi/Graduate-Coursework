```r
data <- data.frame(minutes_swam = c(35, 40, 45, 50, 55),
                   daily_calories = c(450, 500, 550, 600, 650))

plot(data$minutes_swam, data$daily_calories,
     xlab = "Minutes Swam", ylab = "Daily Calorie Consumption",
     main = "Relationship between Minutes Swam and Daily Calorie Consumption")

model <- lm(daily_calories ~ minutes_swam, data = data)

cat("Intercept:", coef(model)[1], "\n")
cat("Coefficient:", coef(model)[2], "\n")

abline(model, col = "lavender")
```
