```r
sample_incomes <- c(97.82138, 98.67651, 97.80384, 97.92021, 98.81304,
                    97.96004, 98.26793, 98.97997, 97.46191, 99.37323,
                    99.67803, 98.70392, 98.57253, 97.94584, 98.61963,
                    97.94205, 98.48125, 97.81605, 98.82261, 98.75639,
                    99.13282, 99.63726, 97.84347, 98.94393, 97.31063,
                    97.97241, 98.92979, 97.78072, 98.52047, 97.67014,
                    98.03187, 98.25676, 97.86675, 96.95675, 98.45049,
                    98.60867, 98.37789, 97.95108, 98.89890, 99.01991,
                    96.66481, 99.05201, 97.07024, 98.04948, 97.52770,
                    97.85462, 99.18324, 98.64935, 98.62164, 96.86658,
                    98.41857, 98.71370)

sample_incomes
```

```r
sample_mean <- mean(sample_incomes)
sample_sd <- sd(sample_incomes)

cat(" mean: ", sample_mean, "\n", "standard deviation: ", sample_sd)
```

```r
conf_level <- 0.98

t_value <- qt(1 - ((1 - conf_level) / 2), df = length(sample_incomes) - 1)

margin_of_error <- t_value * sample_sd / sqrt(length(sample_incomes))

lower_bound <- sample_mean - margin_of_error
upper_bound <- sample_mean + margin_of_error

cat(" confidence interval lower bound: ", lower_bound, "\n", "confidence interval upper bound: ", upper_bound)
```

```r
hist(sample_incomes, main = "Histogram of annual income of electrical and computer graduates", xlab = "Income", xlim = c(95, 101))
abline(v = sample_mean, col = "black")
abline(v = lower_bound, col = "red")
abline(v = upper_bound, col = "red")
```
