### part 3: pie chart

```r
grd = table(data$grade)
grades = as.vector(names(grd))
grd_freq = as.vector(as.numeric(grd))
slices_labels <- round(grd_freq/sum(grd_freq) * 100, 1)
slices_labels <- paste(slices_labels, "%", sep="")
pie(grd, main="frequency of grades", col=rainbow(length(grades)), labels=slices_labels, cex=1.2)

legend(-1.2,.5, grades, fill = rainbow(length(grades)))
```

### part 4: boxPlot

```r
boxplot(data$ts ~ data$sex,
main = "Total Score Distribution",
xlab = "gender", ylab = "Total Score (TS)",
col = rainbow(length(unique(data$sex))),
notch = TRUE, varwidth=TRUE)
```
